```python
R0 = 5.
nu = 1/14. # 1/days
beta = R0*nu
```

```python
S0 = 60e6
I0 = 1.
R0 = 0
```

```python
s_list = [S0]
i_list = [I0]
r_list = [R0]

s, i, r = S0, I0, R0
N = s+i+r
for _ in range(365):   
    delta_s = -beta * i/N * s
    delta_i = + beta * i/N * s - nu*i
    delta_r = + nu*i
    s = s+delta_s
    r = r+delta_r
    i = i+delta_i
    print(delta_i)
    s_list.append(s)
    r_list.append(r)
    i_list.append(i)
```

```python
print(s_list)
print(r_list)
print(i_list)
```
